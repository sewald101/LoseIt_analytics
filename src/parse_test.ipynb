```python
def process_collections(coll):
    """Iterates over Mongodb collections of Lose It forum pages to
    parse and insert messages into sql
    INPUT:
     -- coll: 
    """
    psy2_conn = psycopg2.connect(dbname='loseit', host='localhost')
    psy2_cur = psy2_conn.cursor()
    insert_q = insert_query('posts')
    counter = 0
    for doc in coll:
        insert_posts_to_sql(doc, psy2_cur, insert_q)
        counter += 1
        print 'Documents processed: {}'.format(counter)
        if counter % 50 == 0:
            psy2_conn.commit()
    psy2_conn.commit()
    print "\nCOLLECTION PROCESSED\n"
    psy2_conn.close()
```

```python
process_collections(diet_docs)
process_collections(fitness_docs)
```

```python
def write_topic_urls(library, mongo_db, collection, write_file):
    """From LoseItCatalogParser.topics_lst, inserts topics into psql
    and writes topic landing pg urls to file (f)
    
    PARAMETERS:
    -- library: name of library from with documents scraped (string)
    -- mongo_db: mongo database name (string)
    -- collection: mongo db collection name (string)
    -- write_file: path/file.txt to which topic urls are written (string)  
    """
    mc = MongoClient()
    db = mc[mongo_db]
    coll = db[collection]
    
    with open(write_file, 'w') as f:
        for doc in coll.find({}):
            libr_pg_parsed = LoseItCatalogParser(library, doc['html'], doc['url'])
            libr_pg_parsed.parse_page()
            for topic in libr_pg_parsed.topic_lst[:10]:
                f.write(topic['landing_url'] + '\n')
```

```python
library = 'Fitness and Exercise'
mongo_db = 'lose_it'
coll = 'fitness_ex_catalog'
wfile = '../data/urls_topics_fit_ex.txt'


write_topic_urls(library, mongo_db, coll, wfile)
```

```python
def insert_topics_sql(library, mongo_db, collection, psql_db=None, table=None):
    """From LoseItCatalogParser.topics_lst, inserts topics into psql
    and writes topic landing pg urls to file (f)
    
    PARAMETERS:
    -- library: name of library from which documents were scraped (string)
    -- mongo_db: mongo database name (string)
    -- collection: mongo db collection name (string)
    -- db: psql database (string)
    -- table: into which records are inserted (string)   
    """
    mc = MongoClient()
    db = mc[mongo_db]
    coll = db[collection]
    
    with psycopg2.connect(dbname=psql_db, host='localhost') as conn:
        with conn.cursor() as curs:
            for doc in coll.find({}):
                libr_pg_parsed = LoseItCatalogParser(library, doc['html'], doc['url'])
                libr_pg_parsed.parse_page()
                for topic in libr_pg_parsed.topics_lst:
                    f.write(topic['landing_url'])
                    
    conn.close()
```
